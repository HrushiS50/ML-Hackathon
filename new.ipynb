```python

from PIL import Image

# Define a function to preprocess an individual image
def preprocess_image(image_path, target_size=(224, 224)):
    """
    Preprocess a single image by resizing, converting to a NumPy array, and normalizing pixel values.
    
    Args:
        image_path (str): The file path to the image.
        target_size (tuple): The target size to resize the image (width, height).
        
    Returns:
        numpy.ndarray: Preprocessed image as a NumPy array.
    """
    # Open the image
    image = Image.open(image_path)

    # Resize the image to the target size
    image = image.resize(target_size)

    # Convert the image to a NumPy array
    image_array = np.array(image)

    # Normalize the pixel values to the range [0, 1]
    image_array = image_array / 255.0

    return image_array

# Function to preprocess all images in a directory
def preprocess_images_from_directory(directory, target_size=(224, 224)):
    """
    Preprocess all images in a directory by resizing, converting to NumPy arrays, and normalizing.
    
    Args:
        directory (str): Directory containing image files.
        target_size (tuple): The target size to resize the image (width, height).
        
    Returns:
        numpy.ndarray: Array containing all preprocessed images.
    """
    preprocessed_images = []
    image_filenames = []

    # Loop through all files in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.jpg') or filename.endswith('.png'):  # Only process images with specific extensions
            image_path = os.path.join(directory, filename)

            # Preprocess the image and append it to the list
            preprocessed_image = preprocess_image(image_path, target_size)
            preprocessed_images.append(preprocessed_image)
            image_filenames.append(filename)

    # Convert the list of preprocessed images into a NumPy array
    preprocessed_images = np.array(preprocessed_images)

    return preprocessed_images, image_filenames

# Example usage
image_directory = r'C:\Users\Hrushikesh\OneDrive\Desktop\ML Hackathon\student_resource 3\downloaded_images_2\downloaded_images_2'  # Replace with your directory
preprocessed_images, image_filenames = preprocess_images_from_directory(image_directory)

# Print information about the preprocessed images
print(f"Number of preprocessed images: {len(preprocessed_images)}")
print(f"Shape of each image: {preprocessed_images[0].shape}")  # e.g., (224, 224, 3)

```
